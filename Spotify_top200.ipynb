```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from datetime import datetime
```

```python
# select the date for starting and ending of top200 track 

datelist = pd.date_range(start='2021-02-14', end='2021-02-15')

dates = datelist.strftime("%Y-%m-%d").tolist()

#for having top 200 in any country copy abbreviation in the countries list 

countries =["global","us"] 

top200=[]

for country in countries:
    
    for date in dates:

        url = f"https://spotifycharts.com/regional/{country}/daily/{date}"
        response = requests.get(url)
        soup = BeautifulSoup(response.content,"html.parser")

        for rank,track in enumerate(soup.find_all("td",class_ ="chart-table-track"),1):
            
            title = track.find("strong")
            
            artist=track.find("span")
            
            streams = soup.find_all("td",class_ ="chart-table-streams")
            
            select = soup.find_all("li",class_="selected")
            
            top200.append({"Rank":rank,"Title": title.text,"Artist":artist.text.strip("by "),"Streams": streams[rank-1].text,"Country":select[0].text,"Date":select[2].text})
```

```python
df = pd.DataFrame(top200)
df.head()
```

```python
df.tail()
```

```python
df["Date"]=pd.to_datetime(df["Date"])
df['Streams'] = df['Streams'].apply(lambda x:x.replace(",","")).astype(str).astype(int)
```

```python
df.to_csv("./spotify_top200.csv")
```
