```python
import requests 
import pandas as pd
from bs4 import BeautifulSoup
```

```python
numbers = [1,51]

Movies,Meta,Gross=[],[],[]

for number in numbers:
    
    url= f"https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start={number}"

    response = requests.get(url)

    soup = BeautifulSoup(response.content,"html.parser")


    for movie in soup.find_all("div",class_ = "lister-item-content"):

                                       
        rank = movie.find("span" ,class_="lister-item-index unbold text-primary").text.strip(".")
        
        title = movie.find("a").text

        year = int(movie.find("span",class_="lister-item-year text-muted unbold").text.strip("(I) ()"))

        duration = int(movie.find("span",class_="runtime").text.strip(" min"))

        genre = movie.find("span", class_="genre").text.strip("'\n ")

        rate = movie.find("strong").text

                
        Movies.append({"Rank":rank,"Title":title ,"Duration(min)":duration,"Genre":genre,"Rate":rate,
                       "Year":year})

    for m in soup.find_all("span",class_="metascore favorable"):
        
        metascore = int(m.text.strip("      "))
        
        Meta.append({"Metascore":metascore})
   
        
    for g in soup.select('.sort-num_votes-visible span:nth-child(5)'):
        
        gross = str(g.text).strip("$M")
        
        Gross.append({"Gross($M)":gross})
```

```python
df_movie = pd.DataFrame(Movies)
df_meta = pd.DataFrame(Meta)
df_gross = pd.DataFrame(Gross)
```
